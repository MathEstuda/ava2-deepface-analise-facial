```python



import cv2
from google.colab.patches import cv2_imshow
from deepface import DeepFace

# ✅ 1. Carregar imagem
img = cv2.imread("/content/fotofeliz.jpeg")

print("✅ Imagem carregada com sucesso!")
cv2_imshow(img)

# ✅ 2. Analisar com DeepFace
result = DeepFace.analyze(img, actions=['age','gender','emotion','race'])

# ✅ Pegar resultado principal
r = result[0]

# ✅ Traduções
traducao_emocoes = {
    'angry': 'Raiva',
    'disgust': 'Nojo',
    'fear': 'Medo',
    'happy': 'Feliz',
    'sad': 'Triste',
    'surprise': 'Surpresa',
    'neutral': 'Neutro'
}

traducao_raca = {
    'asian': 'Asiático',
    'indian': 'Indiano',
    'black': 'Negro',
    'white': 'Branco',
    'middle eastern': 'Oriente Médio',
    'latino hispanic': 'Latino/Hispânico'
}

# ✅ Aplicando tradução
emocao_pt = traducao_emocoes.get(r['dominant_emotion'], r['dominant_emotion'])
raca_pt = traducao_raca.get(r['dominant_race'], r['dominant_race'])
genero_pt = "Masculino" if r['dominant_gender'] == "Man" else "Feminino"

# ✅ Mostrar resultado FINAL (em português)
print("\n🔎 RESULTADO DA ANÁLISE:\n")

print("Idade estimada:", r['age'])
print("Gênero predominante:", genero_pt)
print("Emoção predominante:", emocao_pt)
print("Raça predominante:", raca_pt)

```
